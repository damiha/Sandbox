```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_vars = 10000
uniform_vars = np.random.random(n_vars)

plt.title("Probability densitify function of a uniform distribution")
plt.hist(uniform_vars, density=True)
plt.show()
```

Wir wollen eine Reihe von Zufallswerten generieren, die alle (negativ) exponentialverteilt sind
ABER wir koennen mit dem PC nur Zufallsvariablen erzeugen, die gleichförmig (uniform) auf [0, 1] verteilt sind

Wahrscheinlichkeitsdichtefunktion der negativen Exponentialverteilung:

$p(x) = \lambda e^{-\lambda x}$

Verteilungsfunktion der negativen Exponentialverteilung:

$P(x) = 1 - e^{-\lambda x}$

Um Zufallszahlen zu generieren, die nicht-uniform verteilt sind, muss man den Wert der Zufallsvariable danach in die inverse der Verteilungsfunktion $P^{-1}(x)$ weiterleiten

Inverse: $P^{-1}(x) = \frac{-1}{\lambda} ln(1-x)$

```python
def generate_exponential_vars(uniform_vars, lambda_val):    
    return (-1.0 / lambda_val) * np.log(1.0 - uniform_vars)
```

```python
def plot_exp_distribution(uniform_vars, lambda_val):
    exp_vars = generate_exponential_vars(uniform_vars, lambda_val)
    plt.title(f"(Negative) exponential distribution with lambda = {lambda_val}")
    plt.hist(exp_vars, density=True, bins=100)
    
    # plotte die theoretische funktion
    xs = np.linspace(0, 10, 100)
    plt.plot(xs, lambda_val * np.exp(-lambda_val * xs), label="neg. exp pdf")
    plt.legend()
    plt.show()
    
    # schätze den Erwarungswert
    print(f"E(x) ~= {np.mean(exp_vars)}")
```

Mit der Exponentialverteilung lässt sich die Zeit zwischen Betriebsstörung von Haushaltsgeräten etc. modellieren
$\lambda$ gibt an, wie viele Störungen man in einem Zeitintervall zu erwarten hat (wie bei der Poisson-Verteilung die Anzahl der Katastrophen). Der Erwartungswert ist mit $E(x) = \frac{1}{\lambda}$ angegeben und beschreibt, wie viel Zeit im Durchschnitt bis zur Betriebsstörung auftritt. Je mehr Störungen pro Intervall (höheres $\lambda$, desto geringer die Zeit $E(x)$ bis die erste Störung auftritt

```python
plot_exp_distribution(uniform_vars, 1.0)
```

```python
plot_exp_distribution(uniform_vars, 2.0)
```

```python
plot_exp_distribution(uniform_vars, 0.5)
```
