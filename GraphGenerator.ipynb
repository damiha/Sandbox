```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import itertools
import random

def generate_fully_connected_weighted(n_nodes):
    nodes = range(n_nodes)
    
    edges = list(itertools.product(nodes, nodes))
    
    # remove loops
    loops = []
    
    #remove duplicates
    duplicates = []
    
    
    for (from_node, to_node) in edges:
        if from_node == to_node:
            loops.append((from_node, to_node))
        
        if (from_node, to_node) not in duplicates:
            duplicates.append((to_node, from_node))
    
    for loop in loops:
        edges.remove(loop)
        
    for duplicate in duplicates:
        if duplicate in edges:
            edges.remove(duplicate)
        
    weighted_edges = []
    
    for e in edges:
        weighted_edges.append((e[0], e[1], random.randint(5, 20)))
    
    return weighted_edges
```

```python

def draw_graph(weighted_edges):
    graph = nx.Graph()

    graph.add_weighted_edges_from(weighted_edges)

    pos = nx.spring_layout(graph, seed = 7)
    
    edge_labels = dict()
    
    for (from_node, to_node, weight) in weighted_edges:
        edge_labels[(from_node, to_node)] = weight

    plt.title("Fully connected weighted graph")
    nx.draw_networkx_nodes(graph, pos)
    nx.draw_networkx_labels(graph, pos, font_size =  10)
    nx.draw_networkx_edges(graph, pos)
    nx.draw_networkx_edge_labels(graph, pos, edge_labels = edge_labels)
    plt.show()

    mst_solution = nx.minimum_spanning_tree(graph)
    mst_weight_sum = 0
    
    for edge in mst_solution.edges(): 
        mst_weight_sum += edge_labels[edge]
    
    print("MST: ", mst_solution.edges())
    print("MST weight sum has value: ", mst_weight_sum)
```

```python
def write_to_file(weighted_edges, file_path):
    
    file = open(file_path, "w")
    
    for (from_node, to_node, weight) in weighted_edges:
        file.write(f"{from_node} {to_node} {weight}\n")
        
    file.close()
    print("Wrote successfully to: ", file_path)
```

```python
weighted_edges = generate_fully_connected_weighted(n_nodes = 5)
print("Generated ",len(weighted_edges), " edges.")

draw_graph(weighted_edges)

write_to_file(weighted_edges, file_path = "mst_1.txt")
```
