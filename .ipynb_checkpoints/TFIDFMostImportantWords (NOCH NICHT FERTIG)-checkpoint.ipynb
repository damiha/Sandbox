```python
# analyse der inauguration speech von obama, bush, clinton

obama_text = ""
clinton_text = ""
bush_text = ""

with open("Obama Inauguration Speech.txt") as f:
    obama_text = f.read().replace("\n", "")

with open("Bush Inauguration Speech.txt") as f:
    bush_text = f.read().replace("\n", "")
    
with open("Clinton Inauguration Speech.txt") as f:
    clinton_text = f.read().replace("\n", "")
```

```python
obama_words = obama_text.split(" ")
clinton_words = clinton_text.split(" ")
bush_words = bush_text.split(" ")
```

```python
def remove_symbols(word):
    
    excluded_symbols = ["\xa0", "(", ")", "[", "]", "--"]
    
    for excluded_symbol in excluded_symbols:
        word = word.replace(excluded_symbol, "")
    
    return word

def split_symbols(word):
    
    split_symbols = [".", ",", ";", ":"]
    
    # TODO recursive splitting
    """
    words = [word]
    
    for split_symbol in split_symbols:
        
        for word_splitted in words:
            
        if split_symbol in word:
            words += word.split(split_symbol)
    
    """
    return words

def flat_map(f, l):
    res = []
    
    for item in l:
        res += f(item)
        
    return res

def preprocessing(words):
    
    # remove punctuation
    words = list(map(remove_symbols, words))
    
    print(len(words))
    
    # split symbols
    words = list(flat_map(split_symbols, words))
    
    print(len(words))
    
    # filtere leere worte heraus
    words = list(filter(lambda w : len(w) > 0, words))
    
    print(len(words))
    
    # in lower case umwandeln
    words = list(map(lambda w : w.lower(), words))
    
    return words
```

```python
obama_words = preprocessing(obama_words)
clinton_words = preprocessing(clinton_words)
bush_words = preprocessing(bush_words)
```

```python
def get_word_count(words):
    
    word_count = dict()
    
    for word in words:
        word_count[word] = word_count.get(word, 0) + 1
    
    return word_count
```

```python
obama_count = get_word_count(obama_words)
clinton_count = get_word_count(clinton_words)
bush_count = get_word_count(bush_words)

documents = [obama_count, clinton_count, bush_count]
```
