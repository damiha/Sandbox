```python
# analyse der inauguration speech von obama, bush, clinton
files_to_load = [
    "Clinton Inauguration Speech.txt",
    "Clinton Inauguration Speech 1997.txt",
    "Bush Inauguration Speech.txt",
    "Bush Inauguration Speech 2005.txt",
    "Obama Inauguration Speech.txt",
    "Obama Inauguration Speech 2.txt"
]

texts = dict()

for file_name in files_to_load:

    with open(file_name) as f:
        texts[file_name.replace(".txt", "")] = f.read().replace("\n", "")
        f.close()
```

```python
words = dict()

for text_name in texts.keys():
    words[text_name] = texts[text_name].split(" ")
```

```python
def remove_symbols(word):
    
    excluded_symbols = ["\xa0", "(", ")", "[", "]", "--"]
    
    for excluded_symbol in excluded_symbols:
        word = word.replace(excluded_symbol, "")
    
    return word

def split_symbols(word):
    
    split_symbols = [".", ",", ";", ":"]
    
    for split_symbol in split_symbols:
        word = word.replace(split_symbol, " ")
    
    # could potentially add empty words
    return word.split(" ")

def flat_map(f, l):
    res = []
    
    for item in l:
        res += f(item)
        
    return res

def preprocessing(words):
    
    # remove punctuation
    words = list(map(remove_symbols, words))
    
    # split symbols
    words = list(flat_map(split_symbols, words))
    
    # filtere leere worte heraus
    words = list(filter(lambda w : len(w) > 0, words))
    
    # in lower case umwandeln
    words = list(map(lambda w : w.lower(), words))
    
    # diese worte wurden nicht gesagt, sondern sind "Regieanweisungen"
    added_words = ["applause", "-", "_"]
    
    words = list(filter(lambda w: w not in added_words, words))
    
    return words
```

```python
processed_words = dict()

for words_name in words.keys():
    processed_words[words_name] = preprocessing(words[words_name])
```

```python
def get_word_count(words):
    
    word_count = dict()
    
    for word in words:
        word_count[word] = word_count.get(word, 0) + 1
    
    return word_count
```

```python
word_counts = dict()

for words_name in processed_words.keys():
    word_counts[words_name] = get_word_count(processed_words[words_name])

documents = word_counts.values()
```

```python
import math

# das ist immer eine zahl >= 0
# log kann niemals negativ werden, denn dafuer muesste zaehler kleiner als nenner werden
# aber nenner immer <= zaehler da, dokumente wo vorkommt immer untermenge von dokumente
def tfidf(word, document, documents):
    
    n_words_in_document = sum(document.values())
    term_frequency = document.get(word, 0) / n_words_in_document
    
    n_documents_with_word = sum([1 if word in d.keys() else 0 for d in documents])
    inverse_document_frequency = len(documents) / n_documents_with_word
    
    #print(f"tf: {term_frequency}, idf: {inverse_document_frequency}")
    return term_frequency * math.log(inverse_document_frequency)
```

```python
def get_tfidf_scores(document, documents):
    
    tfidf_scores = dict()
    
    for word in document.keys():
        tfidf_scores[word] = tfidf(word, document, documents)
        
    return tfidf_scores
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_10_most_important(document, documents, title):
    
    tfidf_scores = get_tfidf_scores(document, documents)
    
    sorted_scores = dict(sorted(tfidf_scores.items(), key = lambda item: item[1], reverse=True)[:10])
    
    xvalues = list(sorted_scores.keys())
    yvalues = list(sorted_scores.values())
    
    plt.title(title)
    plt.barh(xvalues, yvalues)
    plt.show()
```

```python
for document_name in word_counts.keys():
    plot_10_most_important(word_counts[document_name], documents, f"10 most important words of {document_name}")
```
