```python
from numba import njit, vectorize
import numpy as np

# bitte nur den njit decorator benutzen, der verhindert, dass man Python code schreibt
# der nicht zu effizientem machine code kompilieren wuerde (weil z.B. typing verletzt wurde)

# ist in O(nÂ²)
def find_primes(n):
    
    primes = []
    
    for i in range(2, n + 1):
        
        is_prime = True
        for j in range(2, i):
            if i % j == 0:
                is_prime = False
                break
        
        if is_prime:
            primes.append(i)
            
    return primes
            
```

```python
n = 100_000
%time find_primes(n)
```

```python
fast_find_primes = njit(find_primes)

%time fast_find_primes(n)
```

```python
def only_evens(elements):
    
    res = []
    for element in elements:
        
        if element % 2 == 0:
            res.append(element)
    return res
```

```python
n2 = 100000000
elements = list(range(n2))

%time only_evens(elements)
```

```python
fast_only_evens = njit(only_evens)

%time fast_only_evens(elements)

# numba kommt nicht mit python listen zurecht, dann wird eine warnung angezeigt, da von numba hier auch in
# no python mode nicht gut optimiert werden kann
```

```python

elements_np = np.arange(n2)

# der speedup ist hier noch nicht so gross, weil das erste mal kompiliert werden muss
%time fast_only_evens(elements_np)
```

```python
%time fast_only_evens(elements_np)
```

```python
from numba import vectorize

def square_list(elements):
    res = []

    for element in elements:
        res.append(element * element)
        
    return res

@vectorize
def square(x):
    return x * x
```

```python
n3 = 10000000
%time square_list(list(range(n3)))
```

```python
%time square(np.arange(n3))
```

```python
# das zweite mal liegt die kompilierte funktion schon vor
%time square(np.arange(n3))
```
