```python
import random
import matplotlib.pyplot as plt
import numpy as np

# Morris counting
# counting a lot of items implicitely

def regular_counting(elements):
    counter = 0
    
    for element in elements:
        counter += 1
        
    return counter

def morris_counting(elements):
    
    counter = 0
    
    for element in elements:
        coin = random.uniform(0, 1)
        
        if coin <= 2**(-counter):
            counter += 1
            
    return 2**counter - 2
```

```python
limits = [10**z for z in range(1, 6)]

def get_measurements(limits, counts_per_limit):
    
    measurements = np.zeros((len(limits), counts_per_limit))

    for i, limit in enumerate(limits):

        elements = range(limit)

        for j in range(counts_per_limit):
            measurements[i, j] = morris_counting(elements)
            
    return measurements

measurements_10 = get_measurements(limits, 10)
measurements_15 = get_measurements(limits, 15)
measurements_20 = get_measurements(limits, 20)
```

```python
plt.title("Regular counting vs. Morris counting")
plt.plot(limits, np.mean(measurements_10, axis=1), label="Morris counting avg. over 10", linestyle="dashed")
plt.plot(limits, np.mean(measurements_15, axis=1), label="Morris counting avg. over 15", linestyle="dashed")
plt.plot(limits, np.mean(measurements_20, axis=1), label="Morris counting avg. over 20", linestyle="dashed")
plt.plot(limits, limits, label="Regular counting")
plt.legend()
plt.show()
```
