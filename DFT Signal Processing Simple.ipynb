```python
import numpy as np
import matplotlib.pyplot as plt

# so viele samples in einer sekunde
# der parameter veraendert die aufloesung (signal weniger blocky)
n_samples = 100
samples = np.arange(n_samples)

# 1 Hz = 1 Periode / sekunde = 1 / T
f = 1

# hier kann man die laenge des signals einstellen
n_seconds = 1

# wir sind in der TIME DOMAIN (die amplitude ist eine funktion der zeit)
time_steps = (n_seconds * samples) / n_samples
amplitudes = np.sin(2 * np.pi * f * time_steps)


plt.title("Signal 1 Hz")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.plot(time_steps, amplitudes)
plt.show()
```

```python
# Wir wollen von der TIME-DOMAIN, die uns wenig aufschluss ueber den Charakter des Signals gibt
# in die FREQUENCY-DOMAIN wechseln (Amplitude also Anteil abhaengig von der Frequenz)

# die fourier-reihe einer funktion ist die reihendarstellung von sinus und cosinus werten
# d.h. wir koennen eine funktion als gewichtete summe von sinus und kosinus funktionen (mit verschiedenen frequenzen)
# darstellen (also es startet bei k = 1, also 1Hz, dann k = 2 usw.)

# die Fourier-Reihe ist eine Art eine Funktion als Reihe darzustellen (aehnlich wie Taylor-Reihe)

# Wie bekommen wir die Gewichte also die Koeffizienten der Fourier-Reihe raus ??
```

```python
# Die Koeffizienten bekommt man raus, indem man ein Testsignal (einen Sinus oder Cosinus mit einer bestimmten Frequenz)
# an unser Signal (unsere Funktion) dranmultipliziert und dann von -inf bis inf integriert

# Wenn die Ueberlappung / Korrelation (so berechnet man auch Korrelation) gross ist
# dann wird auch das Integral also der resultierende koeffizient gross sein

# Wenn mein signal ein Sinus ist (also ein periodisches signal von -inf bis inf)
# und ich einen sinus ranmultipliziere, bekomm ich auch sin^2(x) und das integral davon konvergiert doch NICHT??

# wir koennen nur die koeffizienten ueber das uneigentliche integral von -inf bis inf berechnen
# wenn |f(x)| auf -inf bis inf konvergiert
```

```python
# Wenn man den Sinus dranmultipliziert, bekommt man den Koeffizient (Anteil) des jeweiligen sinus
# beim ranmultiplizieren des kosinus bekommt man den anteil des kosinus

# wenn man an die funktion e^(-2 pi f)i dranmultipliziert, dann ist das als wuerde man cos(2 pi f)
# und sin(2 pi f) dranmultiplizieren
# Man muss jetzt ein komplexes integral loesen und bekommt eine komplexe zahl raus
# der real teil entspricht dem reelen Cosinus integral, der imaginaer teil dem sinus anteil
# (WIE BEI COMPLEX ANALYSIS LOESEN WIR KOMPLEXE INTEGRALE, UM REELE INTEGRALE ZU LOESEN)
```

```python
# In audio verarbeitung muessen wir zum glueck keine integrale von -inf bis inf berechnen, da jedes signal
# nur eine endliche zeit dauert
# und wir muessen nicht integrieren, da ein sensor nur an gewissen zeitpunkten abtasten kann
# damit wandelt sich das uneigentliche integral in eine endliche summe um
# wir muessen diese summe also die DISKRETE FOURIER TRANSFORMATION berechnen
```

```python
# n = welches sample schauen wir uns an
# N = samples insgesamt (also in unserem signal)
# k = schwingungen uebers gesamte signal
# k/N = Schwingungen pro Sample (sowas wie die Frequenz)
```

```python

# unser signal geht nur 1s, aber wir haben 100 samples
# also ist unsere sampling frequency 100 Hz

# k ist 1 hz, 2hz etc.
def calculate_amplitude(time_steps, amplitudes, k):
    
    n_samples = len(amplitudes)
    recorded_time = np.max(time_steps) - np.min(time_steps)
    
    # da wir nur eine sekunde aufzeichnen
    sampling_frequency = n_samples / recorded_time
    
    # uberhalb der nyquist rate koennen wir keine aussage machen
    if k >= sampling_frequency / 2:
        return 0
    
    N = n_samples
    
    coefficient = 0
    
    for n in range(N):
        
        coefficient += (amplitudes[n] * np.exp(-1j * 2 * np.pi *(k * n / N)))
    
    # der frequency plot ist an der haelfte der sampling frequency gespiegelt
    # bei einer sampling frequency von 100Hz teilt sich der anteil 100% auf k=1 und k = 99 auf
    # da bei k = 50 gespiegelt
    
    # wir verwerfen die oberen und verdoppeln die unteren
    coefficient *= 2
    
    # wir haben summiert und alle werte unseres testsignals (sinus mit 1hz)
    # und unserem aufgezeichneten signal haben uebereingestimmt d.h. wir haben n_samples mal den signal wert aufaddiert
    # wir muessen noch teilen
    # ansonten wuerde anteil von laenge des signals abhaengig werden
    
    return np.absolute(coefficient) / n_samples
```

```python
def get_frequency_coefficients(time_steps, amplitudes, start_bin, end_bin):
    
    n_samples = len(time_steps)
    # wegen nyquist rate
    n_coefficients = int(n_samples / 2)
    
    frequency_coefficients = np.zeros(end_bin - start_bin + 1)
    frequency_bins = np.zeros(end_bin - start_bin + 1)

    
    recorded_time = np.max(time_steps) - np.min(time_steps)
    sampling_frequency = n_samples / recorded_time
    
    # jedes k hat eine soviel hoehere frequency als das letzte
    bin_width = sampling_frequency / n_samples
    
    for k in range(start_bin, end_bin + 1):
        
        frequency_coefficients[k] = calculate_amplitude(time_steps, amplitudes, k)
        frequency_bins[k] = k * bin_width
        
    return frequency_bins, frequency_coefficients
```

```python
freq_bins, freq_coeff = get_frequency_coefficients(time_steps, amplitudes, 0, 10)
print(freq_bins)
print(freq_coeff)
```

```python
def plot_frequencies(frequency_bins, frequency_coefficients, title):
    
    plt.title(title)
    plt.scatter(frequency_bins, frequency_coefficients)
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Amplitude")
    plt.xticks(np.rint(frequency_bins))
    plt.plot()
```

```python
plot_frequencies(freq_bins, freq_coeff, title="Signal 1 Hz (frequency domain)")
```

```python
plt.rcParams["figure.figsize"] = [10, 5]

# ueberlagertes signal und unterschiedlicher zeitraum
n_samples = 1000
# 3 sekunden aufzeichnung
recorded_time = 3.0

time_steps = recorded_time * (np.arange(n_samples) / n_samples)

f1 = 1
f2 = 3
f3 = 5

amplitudes = np.sin(time_steps * 2 * np.pi * f1) + np.cos(time_steps * 2 * np.pi * f2) + np.sin(time_steps * 2 * np.pi * f3)

plt.title("Signal 2Hz, 10Hz, 25Hz")
plt.plot(time_steps, amplitudes)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
freq_bins, freq_coeff = get_frequency_coefficients(time_steps, amplitudes, 0, 30)
plot_frequencies(freq_bins, freq_coeff, title="Frequencies of compound signal")
```
