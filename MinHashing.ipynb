```python
def set_of_k_shingles(sentence, k=2):
    
    sentence = sentence.lower()
    
    punctuation = [".", ",", ";", ":", "(", ")", "[", "]"]
    
    for symbol in punctuation:
        sentence = sentence.replace(symbol, " ")
        
    words = sentence.split(" ")
    words = list(filter(lambda w: len(w) > 0, words))
    
    shingle_set = set()
    
    for i in range(len(words) - k):
        words_in_shingle = words[i: i+k]
        shingle = ""
        
        for j,word in enumerate(words_in_shingle):
            shingle += word
            if j < k - 1:
                shingle += " "
        
        shingle_set.add(shingle)
    
    return shingle_set
```

```python
def jaccard_similarity(first_set, second_set):
    
    intersection = first_set.intersection(second_set)
    union = first_set.union(second_set)
    
    return len(intersection) / len(union)
```

```python
import numpy as np

def input_matrix(sets):
    
    all_elements = set()
    for s in sets:
        all_elements = all_elements.union(s)
    
    all_elements = list(all_elements)
    in_matrix = np.zeros((len(all_elements), len(sets)))
    
    for i, s in enumerate(sets):
        for j, e in enumerate(all_elements):
            if e in s:
                in_matrix[j][i] = 1
                
    return in_matrix
```

```python
first_set = set_of_k_shingles("The sky is blue and the sun is bright")
second_set = set_of_k_shingles("The sun in the sky is bright.")

# das liefert die genau similarity, aber wir wollen es beschleunigen
# indem wir von beiden sets mithilfe von hashfunktionen signaturen generieren
# und man kann zeigen, dass die jaccard-similarity auf den signature ungefaehr die gleiche ist
# d.h. fuer unendlich grosse signaturen sich annaehrt
jaccard_similarity(first_set, second_set)
```

```python
in_matrix = input_matrix([first_set, second_set])
```

```python
import itertools
import random

seed = 31
random.seed(seed)

def signature_matrix_from_permutation(in_matrix, sig_length):
    
    # jede spalte steht fuer ein dokument, laenge einer spalte = sig_length
    signature_matrix = np.zeros((sig_length, len(in_matrix[0])))
    
    # anzahl der elemente im shingle space
    n = len(in_matrix)
    
    # so viele permutationen generieren wie die signatur lang sein soll
    permutations = random.sample(list(itertools.permutations(range(1, n + 1))), k=sig_length)
    
    # jede permutation erzeugt eine zeile
    for i, permutation in enumerate(permutations):
        # loope ueber alle sets, d.h. baue zeile von links nach rechts auf
        for j in range(len(in_matrix[0])):
            
            perm_set = permutation * in_matrix[:, j]
            signature_matrix[i][j] = np.min(perm_set[perm_set > 0])
            
    return signature_matrix
```

```python
# das sind zwei eindimensionale vektoren
# richtig gerechnet nach https://arpitbhayani.me/blogs/jaccard-minhash
def jaccard_similarity_signature(sig_a, sig_b):
    return np.sum(sig_a == sig_b) / len(sig_a)
```

```python
sig_matrix = signature_matrix_from_permutation(in_matrix, 3)
print(sig_matrix)
```

```python
import matplotlib.pyplot as plt

plt.title("How is the jaccard similarity related to the signature length?")

xs = list(range(1, 50))
ys = []

for sig_length in xs:
    in_matrix = input_matrix([first_set, second_set])
    sig_matrix = signature_matrix_from_permutation(in_matrix, sig_length)
    
    sig_a = sig_matrix[:, 0].reshape(-1)
    sig_b = sig_matrix[:, 1].reshape(-1)
    
    ys.append(jaccard_similarity_signature(sig_a, sig_b))
    
plt.plot(xs, ys, label="Jaccard sim on signatures")
plt.axhline(jaccard_similarity(first_set, second_set), linestyle="dashed", label="True jaccard sim")
plt.legend()
plt.show()
```

```python
class RandomHashFunction:
    
    def __init__(self, max_a, max_b):
        self.max_a = max_a
        self.max_b = max_b
        
    def hashf(self, x, n):
        
        random_a = random.randint(1, self.max_a)
        random_b = random.randint(0, self.max_b)
        
        return ((random_a * x) + random_b) % n
```

```python
# um sich sehr genau der jaccard similarity anzunaehern, versucht man lange signaturen zu erzeugen
# (die aber noch signifikant kuerzer sind als die vereinigung aller sets, sonst haette man nichts gewonnen)
# man kann aber wenn man z.B. laenge = 10k zeichen hat, nur schlecht mit einer permutation ueber [10000] arbeiten
def signature_matrix_from_hash_functions(in_matrix, sig_length):
    
    # erstelle so viele hash funktionen wie signaturen
    max_a = 100
    max_b = 100
    h = [RandomHashFunction(max_a, max_b) for i in range(sig_length)]
    
    # wenn man die zeilennummer hat (eine Zahl != 0), dann kann man sie mit allen unterschiedlichen hash funktionen
    # hashen. Jedes mal wird die Zeilennummer auf eine andere "Position" abgebildet. Da die Hashfunktionen
    # unabhaengig sind, stellt das ungefaehr eine permutation dar
    signature_matrix = np.inf * np.ones((sig_length, len(in_matrix[0])))
    
    # zeilennummer (!= 0) mit allen hashfunktionen gehasht h_1(zeile), h_2(zeile)... ~= permutation
    
    # Laufe ueber zeilen = erstelle permutationen
    for i in range(len(in_matrix)):
        
        r = i + 1
        h_r = np.zeros(sig_length)
        
        for j in range(sig_length):
            h_r[j] = h[j].hashf(r, sig_length)
        
        for j in range(len(in_matrix[0])):
            
            if in_matrix[i][j] == 1:
                for k in range(sig_length):
                    signature_matrix[k][j] = min(signature_matrix[k][j], h_r[k])
                
    return signature_matrix
    
```

```python
sig_matrix = signature_matrix_from_hash_functions(in_matrix, 10)
```

```python
import time

plt.title("Does the jaccard similarity between a signature from permutation or from hash function change?")

xs = list(range(1, 200))
from_perm = []
from_hash = []

in_matrix = input_matrix([first_set, second_set])

# wie lange braucht es, die signaturen zu berechnen
start_perm = time.time()

for sig_length in xs:
    sig_perm = signature_matrix_from_permutation(in_matrix, sig_length)
    
    sig_a = sig_perm[:, 0].reshape(-1)
    sig_b = sig_perm[:, 1].reshape(-1)
    
    from_perm.append(jaccard_similarity_signature(sig_a, sig_b))

end_perm = time.time()
start_hash = end_perm

for sig_length in xs:
    sig_hash = signature_matrix_from_hash_functions(in_matrix, sig_length)
    
    sig_a = sig_hash[:, 0].reshape(-1)
    sig_b = sig_hash[:, 1].reshape(-1)
    
    from_hash.append(jaccard_similarity_signature(sig_a, sig_b))
    
end_hash = time.time()
    
plt.plot(xs, from_perm, label="Jaccard sim perm.")
plt.plot(xs, from_hash, label="Jaccard sim hash.")
plt.axhline(jaccard_similarity(first_set, second_set), linestyle="dashed", label="True jaccard sim")
plt.legend()
plt.show()

perm_time = end_perm - start_perm
hash_time = end_hash - start_hash

print(f"Calculating signatures using permutations: {perm_time}")
print(f"Calculating signatures using hashing: {hash_time}")
print(f"Speed up using hashing: {perm_time / hash_time}x")
```
